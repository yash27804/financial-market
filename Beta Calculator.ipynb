```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
name=input("Enter Ticker of Stock:")
years=input("Enter number of years to find beta:")
dependent_data=yf.download(f"{name}.NS",period=f'{years}y',interval='3mo',auto_adjust=False)
```

```python
nifty=yf.download("^NSEI",period=f'{years}y',interval='3mo',auto_adjust=False)
```

```python
dependent_data.columns=dependent_data.columns.get_level_values(0)
nifty.columns=nifty.columns.get_level_values(0)
```

```python
dependent_data=dependent_data[['Adj Close']]
nifty=nifty[['Adj Close']]
```

```python
dependent_data=dependent_data[['Adj Close']].resample('YE').last()
nifty=nifty[['Adj Close']].resample("YE").last()
```

```python
#dependent data
#nifty
```

```python

dependent_data['Return'] = pd.NA 

dependent_data.loc[dependent_data.index[1:], 'Return'] = ((dependent_data['Adj Close'] / dependent_data['Adj Close'].shift(1))-1)*100
```

```python
dependent_data.rename(columns={'Return':'Return(%)'},inplace=True)
```

```python
nifty['Return(%)']=pd.NA
nifty.loc[nifty.index[1:],'Return(%)']=((nifty['Adj Close']/nifty['Adj Close'].shift(1))-1)*100
```

```python
d_array=np.array(dependent_data.loc[dependent_data.index[1:],'Return(%)']).astype(float)
n_array=np.array(nifty.loc[nifty.index[1:],'Return(%)']).astype(float)
```

```python
beta=np.polyfit(d_array,n_array,1)[0]
```
